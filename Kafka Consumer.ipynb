```python
from kafka import KafkaConsumer
```

```python
consumer = KafkaConsumer(
    bootstrap_servers='localhost:9092',
    auto_offset_reset='latest', #Reset partion offsets upon offsetOutOfRangeError
    group_id='test4', #must have a unique consumer group id
    consumer_timeout_ms=10000)
                #how long to listen for message -we do it for 10 seconds
                #because we poll the kafka broker only each couple of hours

consumer.subscribe('tweets-disaster0')
```

```python
for message in consumer:# this acts as "get me an iterator over the latest message i haven'tseen"
    print(message)
```
